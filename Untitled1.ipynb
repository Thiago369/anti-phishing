```python
from PIL import Image, ImageDraw, ImageFont

# Configurações
width, height = 300, 100
text = "Te amo, Podi"
heart = "❤️"
font_size = 30
bg_color = (255, 255, 255)  # Branco
text_color = (255, 0, 0)    # Vermelho

# Criar uma lista para armazenar os frames
frames = []

# Usar uma fonte padrão
try:
    # Tenta usar uma fonte padrão do sistema
    font = ImageFont.truetype("arial.ttf", font_size)
except IOError:
    # Se a fonte padrão do sistema não estiver disponível, use uma fonte básica
    font = ImageFont.load_default()

# Criar frames com o texto e o coração
for i in range(10):  # Número de frames
    frame = Image.new('RGB', (width, height), bg_color)
    draw = ImageDraw.Draw(frame)
    draw.text((10, 10), text, font=font, fill=text_color)
    draw.text((10, 60), heart, font=font, fill=text_color)
    frames.append(frame)

# Salvar como GIF
frames[0].save('/content/te_amo_podi.gif', format='GIF', append_images=frames[1:], save_all=True, duration=500, loop=0)
```
